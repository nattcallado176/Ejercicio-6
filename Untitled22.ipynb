```python
 import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer

breast_cancer = load_breast_cancer()
data = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
data['label'] = breast_cancer.target

# Filtrar los datos por la etiqueta 'malignant' y 'benign'
malignant_data = data[data['label'] == 0]
benign_data = data[data['label'] == 1]

# Crear los histogramas
plt.figure(figsize=(15, 4))

# Histograma de Mean Radius
plt.subplot(1, 4, 1)
plt.hist(malignant_data['mean radius'], alpha=0.5, label='Malignant')
plt.hist(benign_data['mean radius'], alpha=0.5, label='Benign')
plt.xlabel('Mean Radius')
plt.ylabel('Frequency')
plt.legend()
plt.title('Mean Radius')

# Histograma de Mean Texture
plt.subplot(1, 4, 2)
plt.hist(malignant_data['mean texture'], alpha=0.5, label='Malignant')
plt.hist(benign_data['mean texture'], alpha=0.5, label='Benign')
plt.xlabel('Mean Texture')
plt.ylabel('Frequency')
plt.legend()
plt.title('Mean Texture')

# Histograma de Mean Smoothness
plt.subplot(1, 4, 3)
plt.hist(malignant_data['mean smoothness'], alpha=0.5, label='Malignant')
plt.hist(benign_data['mean smoothness'], alpha=0.5, label='Benign')
plt.xlabel('Mean Smoothness')
plt.ylabel('Frequency')
plt.legend()
plt.title('Mean Smoothness')

# Histograma de Mean Symmetry
plt.subplot(1, 4, 4)
plt.hist(malignant_data['mean symmetry'], alpha=0.5, label='Malignant')
plt.hist(benign_data['mean symmetry'], alpha=0.5, label='Benign')
plt.xlabel('Mean Symmetry')
plt.ylabel('Frequency')
plt.legend()
plt.title('Mean Symmetry')

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn. ensemble import RandomForestClassifier
```

```python
breast_cancer = load_breast_cancer()
data = pd.DataFrame(breast_cancer.data, columns = breast_cancer. feature_names)
data["target"] = breast_cancer.target
```

```python
X = data.drop("target", axis=1)
y = data["target"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size= 0.25, random_state=42)
```

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
print(classification_report(y_test, y_pred, target_names=breast_cancer.target_names))
```
