```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Definimos la EDO según la Ley de Enfriamiento de Newton
def newton_cooling(t, T, k, T_a):
    return -k * (T - T_a)

# Parámetros del problema
T0 = 80  # Temperatura inicial del objeto (°C)
T_a = 25  # Temperatura ambiente (°C)
k = 0.1   # Constante de enfriamiento
t_span = (0, 50)  # Intervalo de tiempo (0 a 50 minutos)
t_eval = np.linspace(t_span[0], t_span[1], 500)  # Puntos para evaluar la solución

# Resolviendo la EDO numéricamente
solution = solve_ivp(newton_cooling, t_span, [T0], args=(k, T_a), t_eval=t_eval)

# Graficando los resultados
plt.figure(figsize=(8, 5))
plt.plot(solution.t, solution.y[0], label='Temperatura del objeto (T)')
plt.axhline(y=T_a, color='r', linestyle='--', label='Temperatura ambiente (T_a)')
plt.title('Enfriamiento de Newton')
plt.xlabel('Tiempo (minutos)')
plt.ylabel('Temperatura (°C)')
plt.legend()
plt.grid(True)
plt.show()
```
